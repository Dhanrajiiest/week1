```python
# Fake Job Title Generator
import random

adjectives = ["Global", "Certified", "Innovative", "Dynamic", "Quantum", "Virtual"]
roles = ["Meme", "Blockchain", "Synergy", "Unicorn", "Vibe", "AI"]
suffixes = ["Engineer", "Strategist", "Evangelist", "Ninja", "Consultant", "Guru"]

def generate_fake_title():
    adj = random.choice(adjectives)
    role = random.choice(roles)
    suffix = random.choice(suffixes)
    return f"{adj} {role} {suffix}"

while True:
    title = generate_fake_title()
    print(f"Your fake job title is: {title}")

    cont = input("Would you like another? (yes/no): ").strip().lower()
    if cont != 'yes':
        print("Thanks for playing with fake titles!")
        break
```
